```python
from nba_api.stats.static import players

# using static to search for players to get player_id
nba_players = players.get_active_players()

print('Number of players fetched: {}'.format(len(nba_players)))
nba_players[:5]
```

```python
def get_playerID(Fname, Lname):
    x = [player for player in nba_players
        if player['first_name'] == f'{Fname}' and player['last_name'] == f'{Lname}'][0]
    return x['id']

get_playerID('Bam', 'Adebayo')
```

```python
def get_playergl(Lname, Fname):
    x = [player for player in nba_players
        if player['first_name'] == f'{Fname}' and player['last_name'] == f'{Lname}'][0]
    ID = x['id']
    gl = playergamelog.PlayerGameLog(player_id = ID)
    gl = gl.get_data_frames()[0]
    return gl

get_playergl('Adebayo', 'Bam')
```
