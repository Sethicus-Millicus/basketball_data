```python
def get_playergl(Lname, Fname):
    x = [player for player in nba_players
        if player['first_name'] == f'{Fname}' and player['last_name'] == f'{Lname}'][0]
    ID = x['id']
    gl = playergamelog.PlayerGameLog(player_id = ID)
    gl = gl.get_data_frames()[0]
    return gl

get_playergl('Adebayo', 'Bam')
```

```python
from nba_api.stats.endpoints import fantasywidget

fantasydf = fantasywidget.FantasyWidget(todays_players = 'Y')
fantasydf = fantasydf.get_data_frames()[0]
fantasydf
```

```python
fantasydf.rename(columns={"PLAYER_POSITION": "POS", "TEAM_ABBREVIATION": "TEAM", "FAN_DUEL_PTS":"FD"}, inplace=True)
fantasydf.drop(columns=['PLAYER_ID', 'TEAM_ID', 'NBA_FANTASY_PTS', 'FT_PCT'], inplace=True)
```

```python
fantasydf = fantasydf.sort_values(by=['TEAM'])
fantasydf
```

```python
fantasydf['team_rank'] = fantasydf.groupby(['TEAM'])['FD'].rank(method='max', ascending = False).astype(int)


```

```python
fantasydf = fantasydf[fantasydf['team_rank'] <= 5]
fantasydf 

    
```
